```python
import pandas as pd

```

```python
df=pd.read_csv("C:/Users/ravi/Downloads/german_credit_data.csv")
```

```python
df.head
```

```python
df.dropna(inplace=True)
```

```python
#label encoding {1st method}

from sklearn.preprocessing import LabelEncoder
label_enc=LabelEncoder()
df['Sex']=label_enc.fit_transform(df['Sex'])
df





```

```python
from sklearn.preprocessing import OneHotEncoder
encoder=OneHotEncoder(sparse_output=False)
encoded_data=encoder.fit_transform(df[["Housing"]])
encoded_data
```

```python
df.isnull().sum()
```

```python

one_hot_df=pd.DataFrame(encoded_data,columns=encoder.get_feature_names_out(["Housing"]))
one_hot_df
                         
```

```python
from sklearn.preprocessing import OrdinalEncoder
ord_enc=OrdinalEncoder()
df["Encode_Job"]=ord_enc.fit_transform(df[["Job"]])
```

```python
df_encode=pd.concat([df,one_hot_df],axis=1)
df_encode
```

```python
#savings accounts by ordinal encoding
from sklearn.preprocessing import OrdinalEncoder
ord_enc=OrdinalEncoder()
df["encode_Saving accounts"]=ord_enc.fit_transform(df[["Saving accounts"]])
df
```

```python
#checking accounts by ord enc
from sklearn.preprocessing import OrdinalEncoder
ord_enc=OrdinalEncoder()
df["encode_Checking account"]=ord_enc.fit_transform(df[["Checking account"]])
df
```

```python
#purpose by One Hot encoding
encoder=OneHotEncoder(sparse_output=False)
df["Purpose"]=encoder.fit_transform(df[["Purpose"]])
df
```
